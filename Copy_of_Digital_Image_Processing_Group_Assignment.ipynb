# **Upload a image and preview it** ðŸ”¼

---



```python
from google.colab import files
from PIL import Image
import IPython.display as display

uploaded_image = None

def upload_image():
    global uploaded_image
    uploaded = files.upload()
    for name, data in uploaded.items():
        uploaded_image = Image.open(name)
        display.display(uploaded_image)
        return uploaded_image

uploaded_image = upload_image()
```

# **Image Rotation Feature** â†©

---



```python
from PIL import Image
import matplotlib.pyplot as plt

if uploaded_image is not None:
    angle = int(input("Enter the rotation angle (in degrees): "))

    # Rotate the image
    rotated_image = uploaded_image.rotate(angle)

    # display
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(uploaded_image)
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    axes[1].imshow(rotated_image)
    axes[1].set_title('Rotated Image')
    axes[1].axis('off')
    plt.show()
else:
    print("No image uploaded. Please run the first code block to upload an image.")
```

# **Image Flip Feature** â–¶â—€

---



```python
from PIL import Image
import matplotlib.pyplot as plt

if uploaded_image is not None:
    flip_direction = input("Enter the flipping direction (h for horizontal, v for vertical): ")

    # Flip direction if/else
    if flip_direction == 'h':
        flipped_image = uploaded_image.transpose(Image.FLIP_LEFT_RIGHT)
    elif flip_direction == 'v':
        flipped_image = uploaded_image.transpose(Image.FLIP_TOP_BOTTOM)
    else:
        print("Invalid flipping direction!")

    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(uploaded_image)
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    axes[1].imshow(flipped_image)
    axes[1].set_title('Flipped Image')
    axes[1].axis('off')
    plt.show()
else:
    print("No image uploaded. Please run the first code block to upload an image.")
```

# **Convert to B&W and Grayscale :** âšª âš«

---



```python
from PIL import Image
import matplotlib.pyplot as plt

if uploaded_image is not None:
    bw_image = uploaded_image.convert("1")
    grayscale_image = uploaded_image.convert("L")

    # Display
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(uploaded_image)
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    axes[1].imshow(bw_image, cmap='gray')
    axes[1].set_title('Black and White Image')
    axes[1].axis('off')
    axes[2].imshow(grayscale_image, cmap='gray')
    axes[2].set_title('Grayscale Image')
    axes[2].axis('off')
    plt.show()
else:
    print("No image uploaded. Please run the first code block to upload an image.")
```
